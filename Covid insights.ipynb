```python
fig=plt.figure(figsize=(20,10))
sns.set_style(style="whitegrid")
sns.lineplot(x="State/UnionTerritory",y="Active_cases",data=df)
plt.xticks(rotation=40,horizontalalignment="right",fontsize="x-large")
plt.tight_layout()
plt.show()
```

```python
fig=plt.figure(figsize=(20,10))
sns.set_style(style="whitegrid")#whitegrid
sns.barplot(x="State/UnionTerritory",y="Active_cases",data=df)
plt.xticks(rotation=40,horizontalalignment="right",fontsize="x-large")
plt.tight_layout()
plt.show()
```

```python
fig=plt.figure(figsize=(20,10))
sns.barplot(x="State/UnionTerritory",y="Cured",data=df,palette="tab10")
plt.xticks(rotation=40,horizontalalignment="right",fontsize="x-large")
plt.tight_layout()
plt.show()
```

```python
highest=df.groupby("State/UnionTerritory")[["Cured","Deaths","Confirmed"]].sum().sort_values(by="State/UnionTerritory")
highest
```

```python
fig=plt.figure(figsize=(20,10))
sns.lineplot(x="Deaths",y="Cured",data=highest,palette="")
plt.xticks(rotation=40,horizontalalignment="right",fontsize="x-large")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,10))
sns.set_style(style="whitegrid")
sns.scatterplot(x='Confirmed',y='State/UnionTerritory',hue="Deaths",data = highest,palette="husl")
```

```python
plt.figure(figsize=(15,8))
sns.countplot(data=highest,x='Deaths')
plt.xticks(rotation=90,horizontalalignment="right",fontsize="x-large")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15,8))
sns.boxplot(data=highest,x='Confirmed',y='Cured')
plt.xticks(rotation=90,horizontalalignment="right",fontsize="x-large")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15,8))
sns.swarmplot(data=highest,x='Confirmed',y='Deaths')
plt.xticks(rotation=90,horizontalalignment="right",fontsize="x-large")
plt.tight_layout()
plt.show()
```
